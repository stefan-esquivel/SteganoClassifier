```python
import numpy as np
import pandas as pd
import random
np.random.seed(1337)
random.seed(1337)
from matplotlib import pyplot as plt
import matplotlib
# import tensorflow as tf
# from tensorflow import keras
# from keras.regularizers import l2
# from sklearn.model_selection import KFold
# from matplotlib import pyplot
import ast
from sklearn.linear_model import SGDClassifier
from sklearn import metrics
from sklearn.metrics import confusion_matrix
from sklearn.preprocessing import scale
import pickle
from sklearn.model_selection import train_test_split
```

```python
with open("/kaggle/input/inputdata/X.pickle", 'rb') as pickle_in:
    X = pickle.load(pickle_in)

with open("/kaggle/input/inputdata/y.pickle", 'rb') as pickle_in:
    y = pickle.load(pickle_in)
```

```python
IMG_SIZE = 100
X = np.array(X).reshape(-1, IMG_SIZE * IMG_SIZE)
y = np.array(y).reshape(-1, 1)
```

0 ->> Cover<br>
1 ->> UERD

```python
y_new=np.zeros(y.shape)
```

```python
for iterator in range(y.shape[0]):
    if y[iterator][0]=="Cover":
        y_new[iterator][0]=0
    if y[iterator][0]=="UERD":
        y_new[iterator][0]=1
```

```python
y_new=y_new.astype(int)
```

```python
y_new
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y_new.ravel(), test_size=0.10, random_state=1337)
```

SGDclassifier is a linear classifier that is optimised by stochastic gradient descent. We use hinge loss with l2 penalty to implement linear svm classifier.

```python
clf = SGDClassifier(loss="hinge", penalty="l2")
```

```python
%%time

from sklearn.metrics import log_loss

train_accuracy = []
val_accuracy = []

train_loss = []
val_loss = []

# range = np.arange(1000, 100000, 1000)

range = np.arange(1000, 50000, 1000)

for i in range:
    print(f"Fitting with {i} samples")
    clf = SGDClassifier(loss="hinge", penalty="l2")
    clf.fit(X_train[:i], y_train[:i])
    accuracy = clf.score(X_train,y_train)
    val_acc = clf.score(X_test,y_test)
    loss = log_loss(y_train[:i], clf.predict(X_train[:i]))
    val_l = log_loss(y_test[:i], clf.predict(X_test[:i]))
    train_accuracy.append(accuracy)
    val_accuracy.append(val_acc)
    train_loss.append(loss)
    val_loss.append(val_l)
    csvlogger.info([i, accuracy, loss, val_acc, val_l])
```
