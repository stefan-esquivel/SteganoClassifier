```python
import pandas as pd

df = pd.read_csv('log_abc.csv')
```

```python
df.head()
```

```python
from matplotlib import pyplot as plt

plt.plot(df['samples'], df['accuracy'])
plt.plot(df['samples'], df['val_accuracy'])
plt.title('AdaBoostClassifier Accuracy')
plt.ylabel('accuracy')
plt.xlabel('samples')
plt.legend(['train', 'val'], loc='upper left')
plt.show()
```

```python
plt.plot(df['samples'], df['loss'])
plt.plot(df['samples'], df['val_loss'])
plt.title('AdaBoostClassifier Loss')
plt.ylabel('loss')
plt.xlabel('samples')
plt.legend(['train', 'val'], loc='upper left')
plt.show()
```
